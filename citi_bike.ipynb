```python
# Dependencies & Setup
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Read July 2019 - April 2021 Citi Bike Data Files & Store into Pandas DataFrames
df1 = pd.read_csv("JC-202007-citibike-tripdata.csv", encoding="ISO-8859-1")
df2 = pd.read_csv("JC-202008-citibike-tripdata.csv", encoding="ISO-8859-1")
df3 = pd.read_csv("JC-202009-citibike-tripdata.csv", encoding="ISO-8859-1")
df4 = pd.read_csv("JC-202010-citibike-tripdata.csv", encoding="ISO-8859-1")
df5 = pd.read_csv("JC-202011-citibike-tripdata.csv", encoding="ISO-8859-1")
df6 = pd.read_csv("JC-202012-citibike-tripdata.csv", encoding="ISO-8859-1")
df7 = pd.read_csv("JC-202101-citibike-tripdata.csv", encoding="ISO-8859-1")
df8 = pd.read_csv("JC-202102-citibike-tripdata.csv", encoding="ISO-8859-1")
df9 = pd.read_csv("JC-202103-citibike-tripdata.csv", encoding="ISO-8859-1")
df10 = pd.read_csv("JC-202104-citibike-tripdata.csv", encoding="ISO-8859-1")
```

```python
# Combine Both July DataFrames from 2019-2020 Into Single DataFrame
combined_df = df1.append([df2, df3, df4, df5, df6, df7, df8, df9, df10], ignore_index=True)
combined_df
```

```python

# Save Combined DataFrames to CSV
combined_df.to_csv("07-2020-04-2021-citibike-tripdata.csv", index=False, header=True)
```
